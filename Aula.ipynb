# Aula de NumPy

Importação e instalação do NumPy

```python
import numpy as np
```

Criando um array de uma dimensão

```python
numeros = np.array([1,2,3,4,5,6,7,8], np.int64)
numeros[3] = 90
numeros
```

Criando um array de duas dimensões (bi-dimensional)

```python
m = np.array([[1,3,5], [6,7,8]])
m[0, 2] # Linha e Coluna
m[1, 1] = 90
m
```

Acessando e visualizandos as arrays

```python
m[1, :] # Aqui seria, somente da linha 1, todos os itens das colunas, isso serve para ver a linha inteira
m[:, 0] # Aqui seria, tudo da linha, mas so da coluna 0, isso serve para ver a coluna inteira
```

Descobrir infos do array

```python
m.ndim # Retorna o número de dimesões de um array
m.shape # Retorna o número de linhas e colunas
m.size # Retorna o número total de elementos de uma arrat, multiplicando linhas x colunas
```
