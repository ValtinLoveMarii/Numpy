```python
#EXE21 - Retorne o tipo de clima

temp = np.array([12, 15, 34, 40, -12, -10])
conditions = [
    temp < 0,
    (temp >= 0) & (temp < 20),
    (temp >= 20) & (temp < 30),
]
choices = ['Freezing', 'Good', 'Warm']
weather = np.select(conditions, choices, default='Very Hot')
print(weather)
```

```python
#EXE27 - Multiplique a matriz notas pelo array pesos usando broadcasting de forma que a primeira coluna seja multiplicada por 1, a segunda por 2 e a terceira por 1.
notas = np.array([[8, 7, 9], 
                  [4, 5, 6], 
                  [10, 9, 8]])

peso = np.array([[1, 2, 1]])
media = (notas * peso).sum(axis=1) / 4
print(np.round(media))
```
