```python
import numpy as np
```

```python
#EXE1 - Crie um array 1D com os números de 1 a 10
print(np.arange(1,11))
```

```python
#EXE2 - Crie um array 2D (3x3) só com zeros
print(np.zeros((3,3)))
```

```python
#EXE3 - Crie um array 4x4 só com uns
print(np.arange(1,17).reshape(4,-1))
```

```python
#EXE4 - Crie um array com 5 números igualmente espaçados entre 0 e 1
print(np.linspace(0, 1, num=5, retstep=True, endpoint=True))
```

```python
#EXE5 - Crie um array com 10 números aleatórios entre 0 e 1
print(np.random.randint(1, 20, 10))
```

```python
#EXE6 - Verifique o tipo (dtype) de um array
arry1 = np.array([1,2,3], dtype=float)
print(arry1.dtype)
```

```python
#EXE7 - Altere o tipo de um array de float para int
arry1.astype(int)
```

```python
#EXE8 - Calcule a média por linha e por coluna
arry1 = np.arange(1, 13).reshape(3, -1)
# axis1 = linha e axis0 = coluna
print(arry1.mean(axis=1))
print(arry1.mean(axis=0))
```

```python
#EXE9 - Encontre o maior valor de cada linha
print(f'Maior valor por linha {arry1.max(axis=1)}')
print(f'Maior valor por coluna {arry1.max(axis=0)}')
```

```python
#EXE10 - Some os valores de cada coluna
print(arry1)
arry1.sum(axis=0).sum()
```

```python
#EXE11 - Normalize os dados entre 0 e 1
arry = np.arange(1, 21).reshape(5, -1)
normalizado = (arry - arry.min()) / (arry.max() - arry.min())
print(f'Valor normalizado {normalizado}')
```

```python
#EXE12 - Selecione valores maiores que 10
print(arry[arry >= 10])
```

```python
#EXE13 - Substitua valores negativos por zero
arry = np.array([[-1, 2, 4],[1, -9, 0]])
arry[arry < 0] = 0
print(arry)
```

```python
#EXE14 - Conte quantos valores são maiores que a média
m = arry.mean()
print(m)
arry[arry > m]
```

```python
#EXE15 - Filtre apenas números pares
arry[arry % 2 == 0]
```

```python
#EXE16 - Encontre índices onde os valores são maiores que 50arry = np.random.randint(1, 100, 10).reshape(2, -1)
arry = np.random.randint(5, 100, 10).reshape(5, -1)
print(arry)
mask = np.where(arry > 50)
# print(f'Valores que satisfazem a condição {arry[mask]}')
print(arry[mask])
```

```python
#EXE17 - Verifique se todos os valores são positivos
lista = arry >= 0
np.unique_values(lista)
```

```python
#EXE18 - Verifique se algum valor é negativo
print(arry)
mask = np.where(arry < 0, True, False)
if True in mask:
    print('Possui valor nulo')
else:
    print('Nao possui valor nulo')
```

```python
#EXE19 - Combine duas condições (> 10 e < 50)
print(arry)
mask = np.where((arry >= 10) & (arry <= 50), 100, 0)
mask
```
