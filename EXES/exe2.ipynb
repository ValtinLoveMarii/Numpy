```python
import numpy  as np
```

```python
#EXE1 - Crie um array 3D de forma (2, 3, 4) preenchido com zeros.
print(np.zeros((2,3,4)))
```

```python
#EXE2 - Crie um array 3D de forma (3, 2, 2) preenchido com uns
print(np.ones((3, 2, 2)))
```

```python
#EXE3 - Crie um array 3D com valores aleatórios entre 0 e 1, shape (4, 3, 2)
print(np.random.rand(4,3,2))
```

```python
#EXE4  - Crie um array 3D usando arange com shape (2, 2, 5)
print(np.arange(1,21).reshape((2,2,-1)))
```

```python
#EXE5 - Mostre: número de dimensões, shape, tamanho total (size), tipo dos dados (dtype) altere o dtype para float.
c = np.array([[[1,2],[1,2]],[[1,2],[1,2]]])

print(c.shape)
print(c.size)
print(c.dtype)
c.astype(float)
```

```python
#EXE6 - Considere um array A com shape (3, 4, 5), Acesse o primeiro “bloco” (índice 0 da dimensão 0).
# Acesse o último elemento do array.
# Acesse o elemento na posição [1, 2, 3].
# Acesse toda a linha de índice 1 da matriz de índice 0.

a = np.random.randint(1, 25, size=((3,4,5)))
print(f'Primerio bloco \n {a[0]}')
print(f'Último elemento do array {a[0,0,-1]}')
print(f'Elemento da posição [1,2,3] {a[1, 2, 3]}')
print(f'Primeiro Bloco linha 2 {a[0, 1]}')
print(a)
```

```python
#EXE7 - Selecione apenas os dois primeiros blocos. Selecione todas as linhas e apenas as colunas de índice 2 até 4. Extraia uma submatriz de shape (2, 3, 3)
print(f'Dois primeiros blocos \n {a[0:2]}')
print(f'Todas linhas e apenas colunas 2 a 4 \n {a[:, :, 2:5]}')
print(f'Sub-matrix (2,3,3) \n {a[0:2, 0:3, 0:3]}')
```

```python
#EXE8 - Some 10 a todos os elementos de um array 3D
print(a + 10)
```

```python
#EXE9 - Multiplique todos os valores do bloco 0 por 2
print(a[0] * 2)
```

```python
#EXE10 - Eleve todos os elementos ao quadrado.
print(a ** 2)
```
