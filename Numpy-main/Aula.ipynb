Criando um array de duas dimensões (bidimensional)

```python
m = np.array([[1,3,5], [6,7,8]]) # Aqui ele cria 2 linhas com 3 colunas
z = np.zeros((2, 2)) # Aqui é uma 1 linha 2 colunas, todas vem com Zero, por conta do np.zeros

m[0, 2] # Linha e Coluna
m[1, 1] = 90
```

Descobrir infos do array

```python
m.ndim # Retorna o número de dimesões de um array
m.shape # Retorna o número de linhas e colunas
m.size # Retorna o número total de elementos de uma array, multiplicando linhas x colunas
```

Indexação e Visualização

```python
m[1, :] # Aqui seria, somente da linha 1, todos os itens das colunas, isso serve para ver a linha inteira
m[:, 0] # Aqui seria, tudo da linha, mas so da coluna 0, isso serve para ver a coluna inteira
```

```python
array = np.arange(1, 21).reshape(5, -1)

print(array)

print(array[0, 1]) # Pega a linha 0
print(array[0:2, 0:4]) # Pega da linha 0 a 2 e as colunas da 0 a 4
print(array[:4, 2:]) # Pega tudo até 4 das linhas e da 2 e o resto tudo na coluna 
```

Máscaras

```python
mask = array >= 8  # Ele cria uma verificação, para avaliar os valores, retorna um array ou matriz com true ou false
# para quando passar no array, ele retorne apenas os valores True, que satisfazem a condição

array[mask]

array[(array >= 7) & (array % 2 == 0)]
```

```python
print(array)
mask_where = np.where(array > 15, 0, array) 
# O where ele verifica uma condição e retorna os inidices onde ela é satisfeita
# Ela é usado para criar máscara, onde se pode por numa matriz e retornar os valores que satisfez a condição
# ele retorna 2 tuplas, 1 contem contém quais linhas q tem, e a outra qual a coluna q tem, ent (1, 1, 2), (5, 2, 3) - seria, na linha 1 coluna tal item satifez a condição
mask_where
```

```python
print(array)
# O argwhere tbm verifica uma condição e retorna os indicies, mas não pode ser usado para uma mascara, pois é o indice individual por valor, como se vizesse arr[0,1]
mask_argwhere = np.argwhere(array > 15) 
mask_argwhere
```
